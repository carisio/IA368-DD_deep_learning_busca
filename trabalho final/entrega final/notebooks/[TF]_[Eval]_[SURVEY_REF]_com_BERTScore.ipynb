# Paths

```python
import pickle

with open(f"{survey}", 'rb') as f:
    papers = pickle.load(f)
```

```python
import bert_score
from bert_score import score

refs = []
cands = []

for paper in papers:
    for subsection in paper['subsections']:
        refs.append(subsection['generated_text']['text'])
        cands.append(subsection['survey_text'])
```

# Eval

```python
from bert_score import BERTScorer

scorer = BERTScorer(lang="en")

P, R, F1 = scorer.score(cands, refs)
```

```python
print(f"P: {P.mean()}")
print(f"R: {R.mean()}")
print(f"F1: {F1.mean()}")
```
